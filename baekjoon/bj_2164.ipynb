# No 2164. 카드2(queue)

* 문제

N장의 카드가 있다. 각각의 카드는 차례로 1부터 N까지의 번호가 붙어 있으며, 1번 카드가 제일 위에, N번 카드가 제일 아래인 상태로 순서대로 카드가 놓여 있다.

이제 다음과 같은 동작을 카드가 한 장 남을 때까지 반복하게 된다. 우선, 제일 위에 있는 카드를 바닥에 버린다. 그 다음, 제일 위에 있는 카드를 제일 아래에 있는 카드 밑으로 옮긴다.

예를 들어 N=4인 경우를 생각해 보자. 카드는 제일 위에서부터 1234 의 순서로 놓여있다. 1을 버리면 234가 남는다. 여기서 2를 제일 아래로 옮기면 342가 된다. 3을 버리면 42가 되고, 4를 밑으로 옮기면 24가 된다. 마지막으로 2를 버리고 나면, 남는 카드는 4가 된다.

N이 주어졌을 때, 제일 마지막에 남게 되는 카드를 구하는 프로그램을 작성하시오.

* 입력
    * 첫째 줄에 정수 N(1 ≤ N ≤ 500,000)이 주어진다.
* 출력
    * 첫째 줄에 남게 되는 카드의 번호를 출력한다.


```python
# 수학적인 풀이
# 2의 거듭제곱 중 가장 큰 수가 남는다.
N = int(input())
i = 0
while 1:
    if 2**i <= N < 2**(i+1):
        break
    i+=1

print(2**i)
```

```python
# 정석 풀이 1
N = int(input())
x = list(range(2, N+1))

def func(x):
    if len(x) == 1:
        return x[0]
    else:
        x = x[2:] + x[0:1]
        return func(x)

print(func(x))

# 메모리 초과된 코드
```

```python
# 정석 풀이 2
N = int(input())
x = list(range(2, N+1))

while len(x) > 1:
    x = x[2:] + x[0:1]

print(x[0])

# 시간 초과된 코드
```

```python
# 아이디어 풀이 1
N = int(input())
x = list(range(1, N+1))

def remove(idx, x):
    cnt = 2
    x[idx] = 0
    while cnt > 0:
        idx = (idx + 1)%len(x)
        if x[idx]>0: cnt-=1
    return idx, x

idx = 0
while x.count(0) < N-1:
    idx, x = remove(idx, x)
    print(x)

print(sum(x))

# 시간 초과된 코드
```

```python
# 아이디어 풀이 2
N = int(input())
x = list(range(1, N+1))

# def remove(c, x):
#     if c==0:
#         c = len(x)%2
#         x = [x[i] for i in range(len(x)) if i%2==1]
#     else:
#         x = [x[i] for i in range(len(x)) if i%2==0]
#     return len(x)%2, x

def remove(c, x):
    C = len(x)%2
    X = [x[i] for i in range(len(x)) if i%2==1-c]
    
    return C, X

c = 0
while len(x) > 1:
    c, x = remove(c, x)
    print(x)

print(x[0])

# 시간 초과된 코드
```

```python
# 문제 그대로 쓰기
N = int(input())
x = list(range(1, N+1))
print(x)

def f1(x):
    return x[1:]
def f2(x):
    return x[1:]+x[:1]

while len(x) > 1:
    x = f1(x)
    print(x)
    x = f2(x)
    print(x)
print()
print(x[0])
```

```python
# 아이디어 풀이 3 : 2에서 좀더 시간 단축하기
N = int(input())
x = list(range(1, N+1))

def remove(c, x):
    C = (len(x)+c)%2
    X = x[1-c::2]
    
    return C, X

c = 0
while len(x) > 1:
    c, x = remove(c, x)
    print(x)

print(x[0])

# 성공!!
```
